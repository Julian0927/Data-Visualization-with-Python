## ***Line Plots (Series/Dataframe).***



A line plot or line chart (Gráfico de linea)  consist of:

* A type of plot.

* This plot displays information as a series of data points called 'markers' connected by straight line segments.

* It is a basic type of chart common in many fields.

* We use line plot when you have a continuous data set. 

These are best suited for trend-based (basado en tendencias) visualizations of data over a period of time.


$ \ $

## ***Example (line plot).***

In 2010, Haiti suffered a catastrophic magnitude 7.0 earthquake. The quake caused widespread (generalizada) devastation and loss of life and aout (fuera) three million people were affected by this natural disaster. As part of Canada's humanitarian effort, the Government of Canada stepped up (intensificó) its effort in accepting refugees from Haiti. 

$ \ $

$$\text{Question: Plot a line graph of immigration from Haiti using `df.plot()`.}$$


$ \ $

$(1)$ First, we will extract the data series for Haiti.


````python
#(1) We remember that:
years = list(map(str, range(1980, 2014)))

#(2) dataseries from haiti
haiti = (df_can).loc['Haiti', years] 
haiti.head()
````

$ \ $

$(2)$ Next, we will plot a line plot by appending `.plot()` to the `haiti` dataframe.


````python
haiti.plot()
````

$ \ $

----

## ***Note (previous example).***

*Pandas* automatically populated the x-axis with the index values (years), and the y-axis with the column values (population). However, notice how the years were not displayed because they are of type *string*. 


$ \ $ 

----

## ***Example (line plot).***


Therefore, let's change the type of the index values to *integer* for plotting.

````python
#(1) We remember the definition of haiti
haiti = (df_can).loc['Haiti', years] 

#(2) let's change the index values of Haiti to type integer for plotting
haiti.index = (haiti.index).map(int) 

#(3) We create the plot object haiti
# haiti.plot(kind ='line') = haiti.plot()
haiti.plot(kind ='line')

#(4) We declare the title and the lable of the axis.
plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

#(5) we need this line to show the updates made to the figure
plt.show() 
````

We can clearly notice how number of immigrants from Haiti spiked up (enriquecido) from 2010 as Canada stepped up (intensificó) its efforts to accept refugees from Haiti. 


$ \ $

----

## ***Example (line plot plot).***

We can put some text in the plot.

````python
#(3) We create the plot object haiti
# haiti.plot(kind ='line') = haiti.plot()
haiti.plot(kind ='line')

#(4) We declare the title and the lable of the axis.
plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

#(5) annotate the 2010 Earthquake. 
# syntax: plt.text(x, y, label)
plt.text(2000, 6000, '2010 Earthquake')

#(6) we need this line to show the updates made to the figure
plt.show() 

````
